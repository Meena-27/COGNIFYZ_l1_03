```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Zomato dataset
zomato_data = pd.read_csv("/content/zomato.csv")

# Assuming the dataset has a column 'avg cost (two people)' containing the average cost for two people

# Calculate the percentage of restaurants in each price range category
price_range_counts = zomato_data['avg cost (two people)'].value_counts(normalize=True) * 100

# Sort the price range categories by their average cost
price_range_counts = price_range_counts.sort_index()

# Create a bar graph to visualize the distribution of price ranges among restaurants
plt.figure(figsize=(12, 6))
price_range_counts.plot(kind='bar', color='pink')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Average Cost for Two People')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=45, ha='right')  # Rotate and align labels
plt.tight_layout()
plt.show()
```
